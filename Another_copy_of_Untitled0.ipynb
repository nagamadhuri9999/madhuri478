```python
# step1 - we have to import the data set to the colab
import pandas as pd #pd is a nickname of pandas we have to use pd instead of pandas in full code
df=pd.read_csv('CSV.csv')#df is a variable ,read_csv is a predefined function from pandas package
```

```python
#step2 - grouping the input and output coloumns
#we use iloc
X=df.iloc[:,0:-1]
```

```python
#step3:handling the missing data
#scenerio 1 missing value in numerical coloumn:we have to do mean of remaining or average of the remaining values
#scenerio 2 missing values in non numerical coloumns like names etc:we have to replace the empty place by highly repeated name or value
#senerio  3 missing values in huge data set example 2/20000 rows:we have to remove those empty 2 rows from the data set
import numpy as np
from sklearn.impute import SimpleImputer
#simple imputer is a class from scikitlearn used to find mean mode etc or to assign new values which is empty
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])

```

```python
X
```

```python
# step4 : handling catogorial data means converting text data to numerical data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
#label encoder is a class which was imported from sl.preprocessing library it is used to convert catogorial data to binary data
#coloumntransformer is a class imported from skcompose used to apply changes to coloumns or change numeric data to binary data
#alexa-0,0,1
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')#encoder is a method which converts text to numeric
X=ct.fit_transform(X)
```

```python
X
```

```python
#label encoder is a class which is used to output only
#onehotencoder is applied for input only
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
#step5:split the data into testing and training
from sklearn.model_selection import train_test_split#it is a class imported from sklearn.modelselection used to split data into testing and training
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#step6:feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train[:,3:]=sc.fit_transform(X_train[:,3:])
X_test[:,3:]=sc.transform(X_test[:,3:])
```

```python
#machine learning
#types in machinelearning
#1:supervisioned learning
#is used when our data is model tasks are classification or prediction
#2:unsupervisioned learning
#is used when our data is unlabeled data,model remembers the pattrens,we can perform predictions also.
#3:reinforcementlearning
#it performs tail and error method (model will learn from failures)
#based on task performing and dataset these are chategerised
#logistic regression algorithm=used when model performs classification tasks(tru or false etc only 2 presents)
#linear regression algorithm= have to apply when model performs prediction tasks (tells what is the value values comes in output in bunch of pattren)

```

```python
#step7:applying algorithm to the data
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```
